```python
s = "We tried list and we tried dicts also we tried Zen"

ss = s.split(' ') #공백으로 쪼개기
cnt = {} #기억하기 위한 dictionary(cnt) 초기화
for V in ss:
    if not V in cnt:  # x가 이전에 발견된적이 없다면
        cnt[V] = 1
    else: #그렇지 않다면
        cnt[V] += 1
        #print(V,cnt[V])

cnt
```

```python
for word in cnt:
    print(word,cnt[word])
```
